```python
# loading another dataset
df_country = pd.read_excel('zomatodataset/Country-Code.xlsx')
df_country
```

```python
df.columns
```

```python
# Merge df_country and df on country code
df_final = pd.merge(df, df_country, on='Country Code', how='left')
df_final
```

```python
df_final.Country.value_counts()
```

```python
country_values = df_final.Country.value_counts().values
```

```python
country_names = df_final.Country.value_counts().index
```

```python
# Showing top 3 countries
plt.pie(country_values[:3], labels=country_names[:3], autopct='%1.2f%%')
```

Observation: Zomato maximum records or restaurants are from India after that United states 
    and then United kingdom. So for Zomato major are business are in India

```python
df_final.columns
```

```python
# Now here We will get all the values of rating
df_final_review = df_final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
df_final_review
```

```python
# if we want to change into a dataframe we use reset_index()
df_final_review = df_final_review.reset_index()
df_final_review
```

```python
# To change the name of last columns 0 to country counts
ratings = df_final_review.rename(columns={0: 'Rating count'})
ratings
```

Observations
1. When rating is between 4.5 and 4.9 it indicates Excellent 
2. When rating is between 4.0 and 4.4, it indicates Very Good
3. When rating is between 3.5 and 3.9, it indicates Good
4. When rating is between 2.5 and 3.4, it indicates Average
5. When rating is between 1.8 and 2.4, it indicates Poor
6. When rating is 0.0, it indicated Not rated

```python
plt.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x='Aggregate rating', y='Rating count', data=ratings )
```

```python
sns.barplot(x='Aggregate rating', y='Rating count', data=ratings, hue='Rating color',
            palette=['blue', 'red', 'orange', 'yellow', 'green', 'green'])
```

Observation:

    1. Not rated count is very high
    2. Maximum number of ratings are between 2.5 and 3.4

```python
# Count Plot
sns.countplot(data=ratings, x='Rating color', palette=['blue', 'red', 'orange', 'yellow', 'green', 'green'])
```
